```python
from scipy.stats import pareto
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Pareto Disribution

```python
pare = pareto.rvs(3,size=100)
pare
```

```python
x = np.linspace(0,5,100)
for k in [1,2,3]:
    plt.plot(x, pareto(k).pdf(x))
plt.axvline(1)
    
```

```python
y = np.linspace(0,5,100)
for k in [1,2,3]:
    plt.plot(y, pareto(k).cdf(x))
plt.axvline(1)
```

## Exponential Distribution

```python
from scipy.stats import expon
x = np.linspace(0, 5, 100)
for f in [0.5]:    
    plt.plot(x, expon(f).pdf(x))
#print expon(0.5).pdf(x)
```

```python
x = np.linspace(0,5,100)
for lam in [0.05, 0.10, .15]:
    plt.plot(x, expon(lam).cdf(x))
```
