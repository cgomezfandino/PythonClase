Esta es una forma cuando genero numeros con rango en x: debo de hacer un buque para que me realice
la operacion, ya que no me permite operar elemento a elemento de una lista.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

x1 = range(1,100)
y1 = []
for i in x1:
    y1.append(i**2)
    
print type(x1)
print type(y1)
```

```python
plt.plot(x1,y1);
```

Esta es la segunda forma, cuando genero un array en x con un linspace(), este me permite luego operar directamente
elemento a elemento del array x por un escalar, para obtener y.

```python
x2 = np.linspace(0,1,100)
y2 = x2**2

print type(x2)
print type(y2)
```

```python
plt.plot(x2, y2);
plt.xlabel("Time");
plt.ylabel("Behaviour")
```

Operaciones entre conjuntos

```python
a = set('abracadabra')
b = set('alacazam')
print a #devuelve las letras unicas en a
print a - b #devuelve los elementos que contiene a pero que no esta en b
print a | b #letras que estan en a o en b
print a & b #retorna letras en a y en b
print a ^ b #retorna letras en a o b pero no en ambas
```
