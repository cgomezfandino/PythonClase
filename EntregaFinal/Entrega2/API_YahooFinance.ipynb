## Para Divisas

```python
price = ['Adj Close']
pairs = ['EUR=X','JPY=X','GBP=X']
timeSeries = []

for i in pairs:
    timeSeries.append(web.DataReader(i, 'yahoo')[price])
    
```

```python
divisas = timeSeries[0].copy()
divisas['JPY'] = timeSeries[1]
divisas['GBP'] = timeSeries[2]
divisas.rename(columns={'Adj Close':'EUR'}, inplace=True)
divisas.head()
```
